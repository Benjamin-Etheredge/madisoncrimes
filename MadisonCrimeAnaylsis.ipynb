```python
import pandas as pd

%pylab inline
```

```python
df = pd.DataFrame.from_csv('./output/clean.csv')
df['DateTime'] = pd.to_datetime(df['DateTime'])
df = df.set_index('DateTime')
```

```python
df['week_days'] = df.index.weekday
df['week_of_year'] = df.index.weekofyear
df['month'] = df.index.month
df['hour_of_day'] = df.index.hour
df['ones'] = 1.0
cats = df.groupby('cat').resample('W', how='sum')['ones']
```

```python
df.groupby('week_days').sum().plot(y='ones', legend=False)
```

```python
df.groupby('week_of_year').sum().plot(y='ones', legend=False)
```

```python
df.groupby('month').count().plot(y='ones', legend=False)
```

```python
df.groupby('hour_of_day').count().plot(y='ones', legend=False)
```
